Select the apprioprate features for our task. As I stated above my goal is to identify industries and regions with the most billionaire activity and growth.

We are only concerned about the factors that account for wealth generation. We want to identify the most important features that push the model's insights in the data

```python
data_interest = data[["net_worth","business_category","business_industries","country_of_residence","rank"]]
```

Get numeric part of *net_worth* feature. For every number extracted, put it in a new column called *net_worth_billions* that will replace the *net_worth* column

Fill Null Values

target encode all columns using the year column as the category

Fill the null values in the dataframe

```python
def prepare_data(data_interest):

  data_interest['net_worth_billions'] = data_interest['net_worth'].apply(lambda x: ''.join(char for char in x if char.isdigit()))
  data_interest = data_interest.drop("net_worth", axis="columns")
  for col in data_interest.columns:
    # Calculate the most common value in the column
    most_common_value = data_interest[col].mode()[0]

    # Replace missing values with the most common value
    data_interest[col] = data_interest[col].fillna(most_common_value)
  for col in data_interest.columns:
      if col != "net_worth_billions":
            # Calculate mean target value for each category
            target_means = data_interest.groupby(col)["net_worth_billions"].median()

            # Map target means to the categorical column
            data_interest[col] = data_interest[col].map(target_means)

  return data_interest

data_interest.nunique()
```

```python
data_to_be_used = prepare_data(data_interest)
data_to_be_used.head()
```

```python
# Calculate percentage of null values for each feature
null_percentages = (data_to_be_used.isna().sum() / len(data)) * 100

# Print the results
print("Percentage of null values for each feature: \n {}".format(null_percentages) )

```

# 3. Linear Regression Models


Pipeline for each model

```python
X = data_to_be_used.drop("net_worth_billions", axis="columns")
y = data_to_be_used["net_worth_billions"]
```

```python
# Create a pipeline for Linear Regression
linear_pipeline = Pipeline([
    ('polynomial_features', PolynomialFeatures(degree=2)),

    ('scaler', StandardScaler()),
    ('linear_regression', LinearRegression())
])


param_grid = {
    'linear_regression__fit_intercept': [True, False],  # Include or exclude intercept
    'linear_regression__n_jobs': [1,2,3,4,5]       # Normalize features
}

# Perform grid search and cross-validation
grid_search = GridSearchCV(linear_pipeline, param_grid, cv=5)
grid_search.fit(X, y)


# Print best parameters and score
print("Best Parameters:", grid_search.best_params_)
# print("Best Cross-Validation Score:", grid_search.best_score_)
print("Feature Model:", grid_search.best_estimator_)

best_linear = grid_search.best_estimator_["linear_regression"]
best_linear.fit(X,y)
best_linear.coef_
best_linear.score(X,y)
```

```python
# Create a pipeline for Ridge Regression
ridge_pipeline = Pipeline([
    ('polynomial_features', PolynomialFeatures(degree=2)),
    ('scaler', StandardScaler()),
    ('ridge_regression', Ridge())
])


param_grid = {
    'ridge_regression__alpha': [0.1, 1.0],

}

# Perform grid search and cross-validation
grid_search = GridSearchCV(ridge_pipeline, param_grid, cv=5)
grid_search.fit(X, y)

# Print best parameters and score
print("Best Parameters:", grid_search.best_params_)
# print("Best Cross-Validation Score:", grid_search.best_score_)
print("Feature Model:", grid_search.best_estimator_)

best_ridge = grid_search.best_estimator_["ridge_regression"]
best_ridge.fit(X,y)
best_ridge.coef_
best_ridge.score(X,y)
```

```python
# Create a pipeline for Lasso Regression
lasso_pipeline = Pipeline([
    ('polynomial_features', PolynomialFeatures(degree=2)),
    ('scaler', StandardScaler()),
    ('lasso_regression', Lasso(alpha=1.0))
])

param_grid = {
    'lasso_regression__alpha': [0.1, 1.0],

}

# Perform grid search and cross-validation
grid_search = GridSearchCV(lasso_pipeline, param_grid, cv=5)
grid_search.fit(X, y)

# Print best parameters and score
print("Best Parameters:", grid_search.best_params_)
# print("Best Cross-Validation Score:", grid_search.best_score_)
print("Feature Model:", grid_search.best_estimator_)

best_lasso = grid_search.best_estimator_["lasso_regression"]
best_lasso.fit(X,y)
best_lasso.coef_
best_lasso.score(X,y)
```

```python

# Create a pipeline for Elastic Net with Polynomial Features and Standard Scaler
elastic_net_pipeline = Pipeline([

    ('polynomial_features', PolynomialFeatures(degree=2)),

    ('scaler', StandardScaler() ),
    ('elastic_net', ElasticNet()), # No hyperparameters set here

])

# Parameters for grid search
param_grid = {
    'elastic_net__alpha': [ 0.01, 1.0],
    'elastic_net__l1_ratio':  [ 0.01, 1.0],
}

# Perform grid search and cross-validation
grid_search = GridSearchCV(elastic_net_pipeline, param_grid, cv=5)
grid_search.fit(X, y)

# Print best parameters and score
print("Best Parameters:", grid_search.best_params_)
# print("Best Cross-Validation Score:", grid_search.best_score_)
print("Feature Model:", grid_search.best_estimator_)

best_enet = grid_search.best_estimator_["elastic_net"]
best_enet.fit(X,y)
best_enet.coef_
best_enet.score(X,y)


```

```python
print("ElasticNet Most important factors are: ", list(zip(X.columns, abs(best_enet.coef_))))
print("Lasso Most important factors are: ", list(zip(X.columns, abs(best_lasso.coef_))))
print("Ridge Most important factors are: ", list(zip(X.columns, abs(best_ridge.coef_))))
print("Vanilla regression Most important factors are: ", list(zip(X.columns, abs(best_linear.coef_))))

```
